```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
n=np.arange(-20,41)
def unit_impulse(n):
    return np.where(n==0,1,0)
u=np.heaviside(n,1)
d=unit_impulse(n)
x1=(0.8)**n
x2=(0.9*np.exp(1j*np.pi/10))**n
x3=2*np.cos(2*np.pi*(0.3)*n+np.pi/3)
```

# a)

```python
plt.stem(n,u)
```

# b)

```python
plt.stem(n,d)
```

# c)

```python
plt.stem(n,x1)
```

# d)

```python
plt.stem(n,np.abs(x2),'g',label="abs")
plt.stem(n,np.real(x2),'r',label="real")
plt.stem(n,np.imag(x2),'b',label="imag")
plt.legend()
```

```python
plt.stem(n,np.angle(x2),'p',label="abs")
plt.stem(n,np.imag(x2),'g',label="abs")
plt.stem(n,np.real(x2),'r',label="abs")

plt.legend()
```

```python
a=np.array([1,-1.15,1.5,-0.7,0.25])
b=np.array([0.18,0.1,0.3,0.1,0.18])
N=np.arange(0,101)
t,h = signal.dimpulse((b,a,1),n=len(N))
plt.stem(t,h[0])
```

```python
u=np.ones(101)
y=signal.lfilter(b,a,u)
plt.stem(N,y)
```

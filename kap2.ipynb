```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
x=np.array([1])
y1=np.convolve(np.ones(5),x)
y2=np.convolve(np.array([1,-1,-1,-1,1]),x)
y=np.convolve(np.ones(3),y1+y2)
y1,y2,y
```

```python
u=np.ones(20)
y1=np.convolve(np.ones(5),u)
y2=np.convolve(np.array([1,-1,-1,-1,1]),u)
y=np.convolve(np.ones(3),y1+y2)
y1,y2,y
```

```python
a=np.array([1,-1.15,1.5,-0.7,0.25])
b=np.array([0.18,0.1,0.3,0.1,0.18])
N=np.arange(0,101)
t,h = signal.dimpulse((b,a,1),n=len(N))
plt.stem(t,h[0])
```

# b)

```python
u=np.ones(101)
y=signal.lfilter(b,a,u)
plt.stem(N,y)
```
